# imuops SDK Demo

This notebook shows the public SDK on top of an existing canonical session directory.

```python
from pathlib import Path

from imuops import build_report, load_session, run_audit, run_benchmark, run_replay
from imuops.config import load_defaults
```

```python
cfg = load_defaults()
session_dir = Path('../output/legacy_test1')
print('session exists:', session_dir.exists())
if session_dir.exists():
    session = load_session(session_dir)
    print(session.metadata.model_dump())
```

```python
if session_dir.exists():
    audit_result = run_audit(session, cfg)
    replay_result = run_replay(session, 'madgwick', cfg)
    benchmark_result = run_benchmark(session, session.metadata.task, cfg)
    report_path = session_dir / 'notebook_report.html'
    build_report(session, audit_result, [replay_result], report_path)
    print('trust_score:', audit_result.summary.trust_score)
    print('benchmark:', benchmark_result.summary.model_dump())
    print('report:', report_path)
```
